```python
# Resume Screening and Analysis Project

import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Sample dataset
data = {
    "name": ["Abhishek", "Rahul", "Priya", "Sneha"],
    "resume_text": [
        "Python SQL machine learning data analysis NLP",
        "Java spring boot backend developer microservices",
        "Data science python pandas machine learning deep learning",
        "Excel PowerBI SQL data visualization business analysis"
    ]
}

df = pd.DataFrame(data)

# Text cleaning
def clean_text(text):
    text = text.lower()
    text = re.sub('[^a-zA-Z]', ' ', text)
    return text

df['cleaned'] = df['resume_text'].apply(clean_text)

# Feature extraction
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['cleaned'])

# Job matching
job_desc = ["python machine learning data science"]
job_vec = vectorizer.transform(job_desc)

scores = cosine_similarity(X, job_vec)
df['score'] = scores

df.sort_values(by='score', ascending=False)
```
